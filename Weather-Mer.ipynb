```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv("nycweatherdata.csv")
```

```python
## Change date type to DATETIME to be able to query later

df['DATE'] = df['DATE'].astype('datetime64[ns]')
```

```python
df.head()
```

```python
df.dtypes
```

```python
## Checking to see if we can filter by given year

yr_2013 = df[df['DATE'].dt.year == 2013]
yr_2013
```

```python
## Further filtering to check if we can then filter by month given the year

yr_2013_jan = yr_2013[df['DATE'].dt.month == 1]
```

```python
yr_2013_jan
```
